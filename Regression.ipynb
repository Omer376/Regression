#**Q11)What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?**
#ANS:-
#Heteroscedasticity occurs when the variance of the residuals (errors) is not constant across all levels of the independent variables.In other words, the spread of errors changes with X.
* Unreliable Standard Errors:- Standard errors of regression coefficients become biased.

* Incorrect p-values :-Can lead to wrong conclusions about variable significance (Type I or Type II errors).

* Misleading Confidence Intervals:- Confidence intervals for coefficients may be too wide or too narrow.

* Reduced Efficiency of OLS Estimators:- Coefficients remain unbiased but are no longer the best (minimum variance).

* Problems in Hypothesis Testing:- t-tests and F-tests become invalid or misleading.

* Inconsistent Model Interpretation

#**Q18)- What are the limitations of using R² as a sole measure of model performance?**
#ANS:-
* A high R² shows strong correlation, not a causal relationship between variables.
* R² always increases when more predictors are added, even if they’re irrelevant.It doesn't penalize overfitting.
* A few outliers can inflate or deflate R²
* R² is designed for linear regression.
* R² tells you how well the model explains variation, not how well it predicts unseen data.
* It gives no information about the significance or reliability of individual predictors.


#**Q19)- How would you interpret a large standard error for a regression coefficient?**
#ANS:-
#A large standard error for a regression coefficient means the model is uncertain about that variable’s effect on the outcome. It can be a sign of multicollinearity, noisy data, or weak relationships, and it often leads to insignificant results in hypothesis testing.
# Interpretation of a Large Standard Error:
* The estimate of the coefficient is not precise.
* A large SE often leads to a high p-value, meaning the variable may not be statistically significant.
* If predictors are highly correlated, it inflates the SE of the coefficients.
* A large SE may result from too little data

#**Q28)What are the limitations of polynomial regression?**
#ANS:-
* Can overfit the data, especially with high-degree polynomials
* Performs poorly on unseen data (bad extrapolation)
* Number of terms grows rapidly with more variables and degrees
* Becomes hard to interpret as complexity increases
* Very sensitive to outliers
* Increases computation time with large datasets
* Can cause multicollinearity between polynomial terms

#**Q29)What methods can be used to evaluate model fit when selecting the degree of a polynomial?**
#ANS:-
* Adjusted R² – penalizes for adding unnecessary terms
* Cross-validation – checks model performance on unseen data
* Mean Squared Error (MSE) – lower MSE means better fit
* Root Mean Squared Error (RMSE) – interpretable error metric
* AIC (Akaike Information Criterion) – balances fit and complexity
* BIC (Bayesian Information Criterion) – stronger penalty for complexity

#**Q30)How is polynomial regression implemented in Python?**
#ANS:-


```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import numpy as np
import matplotlib.pyplot as plt

# Sample data (X must be 2D)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25])  # Quadratic relationship

# Define the polynomial degree (e.g., degree 2)
degree = 2

# Create pipeline: Polynomial transformation + Linear Regression
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

# Fit the model
model.fit(X, y)

# Predict
X_test = np.linspace(0, 6, 100).reshape(-1, 1)
y_pred = model.predict(X_test)

# Plot
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X_test, y_pred, color='red', label='Polynomial Fit')
plt.legend()
plt.title(f'Polynomial Regression (degree={degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
